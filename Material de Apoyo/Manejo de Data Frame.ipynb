## .loc y .iloc

Para seleccionar filas y columnas, esta .loc y .iloc

Estas funcionan con df.loc[ filas (lista de nombres) , columnas (lista de nombres) ], si utilizas : selecciona todas las filas o columnas.



```python
df.loc[: # todas las filas
       ,["longitud","latitud"]] # Seleccionar las columnas
```

```python
df.loc[[0, 1] # Seleccionar la fila 0 y 1
       , :] # Seleccionar todas las columnas
```

Con .iloc es diferente, solo recibe numeros en rangos, es decir, 0:10 o una posición especifica.

Recuerden que python empieza desde 0 y cuando se utilizan : el último número no lo toma.

```python
df.iloc[20:30 # selecciona las filas de la posición 20 hasta la 29
        , 2:5 ] # selecciona las columnas de la posición 2 hasta la 5
```

# Filtros

Ahora, en la sección de filtros vamos a considerar diferentes casos.

Podemos seleccionar una columna y realizar diferentes operaciones.

# Operadores logicos con pandas

| **Operador** | **Descripción** | **Ejemplo** |
|--------------|----------------|-------------|
| ==   (igualdad)        | Devuelve True si ambos operandos son iguales, de lo contrario, devuelve False. | `df[columna] == 2`  |
| !=       (Diferente)    | Devuelve True si ambos operandos son diferentes, de lo contrario, devuelve False. | `df[columna] != 2`  |
| > (mayor)            | Devuelve True si el operando izquierdo es mayor que el operando derecho, de lo contrario, devuelve False. | `df[columna] > 1`  |
| <     (menor)       | Devuelve True si el operando izquierdo es menor que el operando derecho, de lo contrario, devuelve False. | `df[columna] < 1`  |
| >=/<= (Mayor/menos o igual) | Devuelve True si el operando izquierdo es mayor o igual/menos o igual que el operando derecho, de lo contrario, devuelve False. | `df[columna] >= 2`  |
| .isin  (Esta en)          | Devuelve True si es un miembro del segundo operando, de lo contrario, devuelve False. | `df[columna].isin([1, 2, 3])` |
| &       (Y)     | Devuelve True si ambos operandos son verdaderos, de lo contrario, devuelve False. | ` (df[columna1] >10) & (df[columna2]<20)`  |
| \|       (O)   | Devuelve True si al menos uno de los operandos es verdadero, de lo contrario, devuelve False. | `(df[columna1] >10) \| (df[columna2]<20)`  |
| ~ (Negación)          | Devuelve False si el operando es verdadero, de lo contrario, devuelve True. | ~(df[columna] <1) |

Con isin es para obtener los que estan en una lista, para el siguiente ejemplo si queremos los que son HERIDO o MUERTO, son varias categorías y queremos filtrar por ambas.

Aplicamos isin y la negación de la siguiente manera, así podemos obtener los registros que no tengan como comuna ("In", "Sin Inf").

Y cómo aplicamos los filtros?

Se puede realizar con [] y .loc[]

```python
df[ df["gravedad"]=="HERIDO" ]
```

Pasamos de 40599 a 23651.



```python
df[ ~df["gravedad"].isin(["HERIDO", "MUERTO"] ) ]
```

También se puede realizar con .loc

```python
df.loc[df["latitud"]>6.21, # Filtro
       :  ] # seleccionando todas las columnas
```

Para guardar los filtros, lo mas apropiado es crear una nueva variable.

```python
df_filtro = df.copy() # El mismo data frame ahora se llama df_filtro
df_filtro = df_filtro.loc[df["latitud"]>6.21,:  ]
df_filtro
```

El filtro quedo guardado en una nueva variable que se llama df_filtro. ¿Y si queremos varios filtros?

Cuando se utiliza el operador &, es recomendable colocar los demás filtros entre parentesis ().


```python
df_filtro = df_filtro.loc[ (df["latitud"]>6.21) & (~df["gravedad"].isin(["HERIDO", "MUERTO"] )) ,:  ]
df_filtro
```

Una forma mas ordenada de realizar filtros podría ser:

```python
df_limpio = df.copy() # El mismo data frame ahora se llama df_limpio
# Crear variables por cada filtro
filtro_gravedad = df_limpio["gravedad"].isin(["HERIDO", "MUERTO"]) # Realizamos el filtro para el tipo de gravedad
filtro_latitud = df_limpio["latitud"]>6.21 # Definimos el filtro para la latitud
filtro_longitud = df_limpio["longitud"]<-75.55 # Definimos el filtro para la longitud
filtro_comuna = df_limpio["comuna"].isin( [ 'In', 'Sin Inf' ] ) # Definimos el filtro para comuna
df_filtro = df_limpio.loc[filtro_gravedad & filtro_latitud & filtro_longitud & ~filtro_comuna, : ] # ahora aplicamos & con todos los filtros
# Al final obtenemos que se cumplan todas las condiciones de los filtros
df_filtro
```

```python
df_filtro["comuna"].value_counts()
```

Ahora vemos que se eliminaron las comunas

```python
df_filtro["gravedad"].value_counts()
```

# Ayuda Reto 2

Filtramos los datos atípicos, luego al filtro le aplicamos la negación.

```python
df_filtro = df.copy() # El mismo data frame ahora se llama df_limpio
# Crear variables por cada filtro
filtro_gravedad = df_filtro["gravedad"].isin(["HERIDO", "MUERTO"]) # Realizamos el filtro para el tipo de gravedad
filtro_atipico = (df_filtro["longitud"]<-75.65) & (df_filtro["latitud"]>6.3) # "atípicos"
filtro_comuna = df_filtro["comuna"].isin( [ 'In', 'Sin Inf' ] ) # Definimos el filtro para comuna
df_filtro = df_filtro.loc[filtro_gravedad & ~filtro_atipico & ~filtro_comuna, : ] # ahora aplicamos & con todos los filtros
# Al final obtenemos que se cumplan todas las condiciones de los filtros
df_filtro
```

# Operaciones entre filas o columnas

Las funciones se llaman: .min(), .max(), .mean(), .idxmax(), .idxmin()

Cada una de estas funciones tiene un argumento axis, axis=1 realiza la operación para filta y axis=0 realiza la operación para columna  


```python
datos_marketing[['MntWines', 'MntFruits',
       'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
       'MntGoldProds']].sum(axis=0 )
```
